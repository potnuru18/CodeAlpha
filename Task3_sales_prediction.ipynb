```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample Sales Data (Replace this with real data)
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct"],
    "Sales": [120, 150, 170, 160, 180, 200, 220, 210, 230, 250]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Define a function for simple moving average prediction
def moving_average(series, n):
    return np.mean(series[-n:])  # Taking the average of the last 'n' months

# Predict next month's sales based on the last 3 months
predicted_sales = moving_average(df["Sales"], 3)

# Print Results
print("=" * 40)
print("Sales Prediction Using Moving Average")
print("=" * 40)
print(f"Last 3 Months Sales Data: {df['Sales'].tolist()[-3:]}")
print(f"Predicted Sales for Next Month: {predicted_sales:.2f}")
print("=" * 40)

# Visualization
plt.figure(figsize=(10, 5))
plt.plot(df["Month"], df["Sales"], marker='o', linestyle='-', color='b', label="Actual Sales")
plt.axhline(y=predicted_sales, color='r', linestyle='--', label="Predicted Sales")

# Customize the graph
plt.xlabel("Month", fontsize=12, fontweight='bold')
plt.ylabel("Sales", fontsize=12, fontweight='bold')
plt.title("Sales Prediction Using Python (Moving Average)", fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True)
plt.show()
```
